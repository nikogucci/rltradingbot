```python
def plot_rl_rewards(selection):
    a = np.load(f'linear_rl_trader_rewards/{selection}.npy')
    print(f"average reward: {a.mean():.2f}, min: {a.min():.2f}, max: {a.max():.2f}")

    if selection == 'train':
      # show the training progress
      plt.plot(a)
    else:
      # test - show a histogram of rewards
      plt.hist(a, bins=20)

    plt.title(selection)
    plt.show()
```

```python
plot_rl_rewards("train")
```

```python
plot_rl_rewards("test")
```
