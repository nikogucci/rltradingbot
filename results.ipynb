```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
```

```python
train_rewards = np.load("linear_rl_trader_rewards/train.npy")
test_rewards = np.load("linear_rl_trader_rewards/test.npy")
random_rewards = np.load("linear_rl_trader_rewards/random.npy")
```

```python
rewards_df = pd.DataFrame({"train":train_rewards, "test":test_rewards, "random":random_rewards})
```

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(20,5))

rewards_df["train"].plot(ax=axes[0])
axes[0].set_title('Portfolio value (train)')
rewards_df["test"].plot(ax=axes[1])
axes[1].set_title('Portfolio value (test)')
rewards_df["random"].plot(ax=axes[2])
axes[2].set_title('Portfolio value (random)')
```

```python
sns.displot(rewards_df)
```

```python
train_rewards_v1 = np.load("linear_rl_trader_rewards/train_v1.npy")
train_rewards_v2 = np.load("linear_rl_trader_rewards/train_v2.npy")

rewards_df = pd.DataFrame({"train_without_ind":train_rewards_v1, "train_with_ind":train_rewards_v2})

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15,5))

rewards_df["train_without_ind"].plot(ax=axes[0])
axes[0].set_title('Portfolio value (train_without_ind)')
rewards_df["train_with_ind"].plot(ax=axes[1])
axes[1].set_title('Portfolio value (train_with_ind)')
```

```python
sns.displot(rewards_df)
```
