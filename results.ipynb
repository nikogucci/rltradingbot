```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
```

```python
def get_rewards_data(versions,indicators):
    rewards_df = pd.DataFrame()
    
    filenames = [v + "_" + i for v in versions for i in indicators]
    
    print(filenames)
    for f in filenames:
        filepath = "linear_rl_trader_rewards/" + f + ".npy"
        reward_data = np.load(filepath)
        single_df = pd.DataFrame({f:reward_data})
       
        rewards_df = pd.concat([rewards_df,single_df], axis=1) 
    return rewards_df

versions = ["test_v1","train_v1"]
indicators = ["appl_msi_sbux_diff","appl_msi_sbux_basic"]

rewards_df = get_rewards_data(versions,indicators)
rewards_df.head()
```
