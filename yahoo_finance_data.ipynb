```python
import matplotlib.pyplot as plt
import numpy as np
import yfinance as yf
import pandas as pd
```

```python
i = "Appl"
indicators = ["Open","Close"]
[i + "_" + str(s) for s in indicators]
```

```python
def get_stock_prices(stocks, indicators):
    df = pd.DataFrame()
    for i in stocks:  # ["AAPL","MSI","SBUX"]:
        stock_df = yf.download(i, start="2015-01-01", end="2020-03-20").copy()
        stock_df = stock_df[indicators].copy()  # stock_df[["Open","Close"]]
        stock_df["Diff"] = stock_df["Close"] - stock_df["Open"]
        print(stock_df.head())
        stock_df.drop(columns=["Open"],inplace =True)
        colnames = [s + "_" + i for s in stock_df.columns]
        print(colnames)
        stock_df.columns = colnames
        df = pd.concat([df, stock_df], axis=1)
    return df


data_df = get_stock_prices(["AAPL", "MSI", "SBUX"], ["Open", "Close"])
data_df = data_df[sorted(data_df.columns)]
print(data_df.head())
```

```python
data_df.to_csv("3_stocks_diff.csv",index=False)
```
